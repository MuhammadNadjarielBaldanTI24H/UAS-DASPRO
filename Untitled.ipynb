```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
df = pd.read_csv("Data Training-1.csv")

print(df.info())
```

```python
X = df[["SLTP"]]
y = df["Universitas"]
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["SLTP"],y=df["Universitas"], label='Data Aktual', color='orange')
sns.lineplot(x=df["SLTP"], y=y_pred, color='red', label='Regresi Linier')
plt.title("Regresi Linier: SLTP vs Universitas")
plt.xlabel("Jumlah Lulusan SLTP")
plt.ylabel("Jumlah Lulusan Universitas")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
last_row = df.iloc[-1]
education_levels = [
    "Tidak/belum pernah sekolah", "Tidak/belum tamat SD", "SD", "SLTP",
    "SLTA Umum/SMU", "SLTA Kejuruan/SMK", "Akademi/Diploma", "Universitas"
]

values = [last_row[col] for col in education_levels]
data_plot = pd.DataFrame({
    'Tingkat Pendidikan': education_levels,
    'Jumlah': values
})
plt.figure(figsize=(12, 6))
sns.barplot(
    data=data_plot,
    x='Tingkat Pendidikan',
    y='Jumlah',
    hue='Tingkat Pendidikan',
    palette='Blues_d',
    legend=False
)
plt.xticks(rotation=45, ha='right')
plt.title(f"Jumlah Penduduk Menurut Pendidikan (Periode {last_row['Periode']} - {last_row['Bulan']})")
plt.ylabel("Jumlah Penduduk")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(values, labels=education_levels, autopct= '%1.1f%%', startangle=140)
plt.title(f"Distribusi Persentase Pendidikan(Periode {last_row['Periode']}-{last_row['Bulan']})")
plt.tight_layout()
plt.show()
```
